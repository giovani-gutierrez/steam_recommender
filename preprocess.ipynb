```python
interactions = pd.read_csv('/kaggle/input/game-recommendations-on-steam/recommendations.csv')
```

```python
interactions_df = interactions.merge(games, on = ['app_id'])
```

```python
interactions_df
```

```python
df2 = interactions_df[['user_id', 'app_id', 'is_recommended', 'hours', 'date',
                       'price_original', 'price_final', 'title', 'date_release',
                       'rating', 'positive_ratio', 'user_reviews']]

df3 = df2.rename(columns = {'user_id': 'userID', 'app_id': 'gameID', 'is_recommended': 'recommended',
                            'hours': 'hours_played', 'date': 'review_date', 'date_release': 'release_date',
                            'rating': 'overall_rating'})
```

```python
df3.info()
```

```python
df3['title'] = df3['title'].astype('string')
df3['overall_rating'] = df3['overall_rating'].astype('category')
df3['release_date'] = pd.to_datetime(df3['release_date'], format = '%Y-%m-%d')
df3['review_date'] = pd.to_datetime(df3['review_date'], format = '%Y-%m-%d')
```
